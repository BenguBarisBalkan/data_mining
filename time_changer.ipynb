```python
import pandas as pd
```

```python
btc = pd.read_csv('./data/btcusd_1-min_data.csv')
```

```python
btc.head()
```

```python
btc['Date'] = pd.to_datetime(btc['Timestamp'], unit='s')
```

```python
btc.head()
```

```python
btc = btc[['Date', 'Open', 'High', 'Low', 'Close']]
```

```python
btc.head()
```

```python
btc.to_csv('./data/btc_minute_2012.csv')
```

```python
btc = btc.set_index('Date')
```

```python
btc_hourly = btc.resample('H').agg(
    {'Open': 'first',
     'High': 'max',
     'Low': 'min',
     'Close': 'last'}
)
```

```python
btc_hourly.head()
```

```python
btc_hourly.to_csv('./data/btc_1hour_2012.csv')
```

```python
btc_4_hourly = btc.resample('4H').agg(
    {'Open': 'first',
     'High': 'max',
     'Low': 'min',
     'Close': 'last'}
)
```

```python
btc_4_hourly.head()
```

```python
btc_4_hourly.to_csv('./data/btc_4hour_2012.csv')
```

```python
btc_daily = btc.resample('1D').agg(
    {'Open': 'first',
     'High': 'max',
     'Low': 'min',
     'Close': 'last'}
)
```

```python
btc_daily.head()
```

```python
btc_daily.to_csv('./data/btc_daily_2012.csv')
```
